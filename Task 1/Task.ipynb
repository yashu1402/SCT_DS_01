```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('dataset.csv')
```

```python
data.isnull().sum()
```

```python
data['age'].fillna(data['age'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
print('Age Distributions:')
print(data['age'].describe())

plt.figure(figsize=(8, 6))
plt.hist(data['age'], bins=15,color='red')
plt.title('Distribution of Age Among Billionaires')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
print('Gender distribution:')
print(data['gender'].value_counts())

plt.figure(figsize=(8,6))
data['gender'].value_counts().plot(kind='bar',color=['lightblue','lightpink'])
plt.title('Distribution of Gender Among Billionaires')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
